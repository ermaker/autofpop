```python
N_COL = 9
N_ROW = 9
def GetCellImage(screenNorm, row, col):
    height, width = screenNorm.shape[:2]
    cellHeight = height / N_ROW
    cellWidth  = width  / N_COL
    y1 = cellHeight * row + (col%2) * cellHeight / 2.0
    y2 = y1 + cellHeight
    x1 = cellWidth * col
    x2 = x1 + cellWidth
    return screenNorm[y1:y2, x1:x2, :]
```

```python

## train model
import recognition
model = recognition.ImgRecognizer()
model.load()
model.train()
```

```python
testim = GetCellImage(scNormalized, 9,1)

print(recognition.CELL_NAMES[model.predict(testim)])
plt.imshow(testim)
```
